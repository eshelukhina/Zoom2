```python
def get_sec_hours(time_str):
    """Get seconds from time."""
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)

def get_sec_minutes(time_str):
    """Get seconds from time."""
    m, s = time_str.split(':')
    return int(m) * 60 + int(s)

def get_seconds(str_time):
    if len(str_time.split(':')) == 2:
        return get_sec_minutes(str_time)
    elif len(str_time.split(':')) == 3:
        return get_sec_hours(str_time)
    else:
        return int(str_time)
```

```python
import statistics
df = pd.read_csv('dataset.csv', sep=';',header=None)

start_volumes = []
end_volumes = []

for i in range(len(df)):
    video_name = 'drive/MyDrive/dataset/' + df.values[i][0]
    print(df.values[i][0])

    start_time = get_seconds(df.values[i][1])
    end_time = get_seconds(df.values[i][2])

    print(start_time, end_time)

    clip = mp.VideoFileClip(video_name)
    audio = clip.audio

    # print(start_time, end_time)

    start_max_volume, end_max_volume = get_threshold(audio, start_time, end_time)

    start_volumes.append(start_max_volume)
    end_volumes.append(end_max_volume)

print("Среднее значение громкости звука для начала лекций", statistics.mean(start_volumes))
print("Медиана громкости звука для начала лекций", statistics.median(start_volumes))

print("Среднее значение громкости звука для окончания лекций", statistics.mean(end_volumes))
print("Медиана громкости звука для окончания лекций", statistics.median(end_volumes))
```
